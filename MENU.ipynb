```python
# LUYỆN TẬP: Thiết kế menu chọn món, hình thức thanh toán(tiền mặt, chuyển khoản giảm 5%), nút thanh toán cho ra giá tiền
from tkinter import *
from tkinter import ttk

def total(event):
    bunbo = int(check1.get())*int(sl1.get())
    bunmoc= int(check2.get())*int(sl2.get())
    phobo = int(check3.get())*int(sl3.get())
    phoga = int(check4.get())*int(sl4.get())
    tinhtien = bunbo+bunmoc+phobo+phoga
    
    discount = int(var1.get())/100
    thanhtoan = tinhtien*(1-discount)
    tongtien['text'] = f'{thanhtoan} VNĐ'
    
def color():
    color=var1.get()
    if color==0:
        tienmat['fg']='red'
        chuyenkhoan['fg']='black'
    else:
        tienmat['fg']='black'
        chuyenkhoan['fg']='red'
        
    

menu=Tk()
menu.title('Menu Quán')
menu.geometry('600x350')
myfont = ('Helvetica','14')
ttfont = ('Helvetica','12')

bar = Label(menu, text='Món ăn', bg='aqua',font=myfont)
bar.place(x=70,y=0)

soluong = Label(menu, text='Số lượng', bg='lightblue',font=myfont)
soluong.place(x=360, y=0)

check1=IntVar()
check2=IntVar()
check3=IntVar()
check4=IntVar()

checkbt1=Checkbutton(menu, text='Bún bò (30k)', onvalue=30000, offvalue=0, variable=check1,font=myfont)
checkbt1.place(x=20, y=30)
checkbt2=Checkbutton(menu, text='Bún mọc (30k)', onvalue=30000, offvalue=0, variable=check2,font=myfont)
checkbt2.place(x=20, y=60)
checkbt3=Checkbutton(menu, text='Phở bò (40k)', onvalue=40000, offvalue=0, variable=check3,font=myfont)
checkbt3.place(x=20, y=90)
checkbt4=Checkbutton(menu, text='Phở gà (40k)', onvalue=40000, offvalue=0, variable=check4,font=myfont)
checkbt4.place(x=20, y=120)

values = [0,1,2,3,4,5]

sl1 = ttk.Combobox(menu,font=myfont, values=values,state='readonly', justify=RIGHT)
sl1.place(x=290,y=30)
sl1.set(values[0])
sl2 = ttk.Combobox(menu,font=myfont, values=values,state='readonly', justify=RIGHT)
sl2.place(x=290,y=60)
sl2.set(values[0])
sl3 = ttk.Combobox(menu,font=myfont, values=values,state='readonly', justify=RIGHT)
sl3.place(x=290,y=90)
sl3.set(values[0])
sl4 = ttk.Combobox(menu,font=myfont, values=values,state='readonly', justify=RIGHT)
sl4.place(x=290,y=120)
sl4.set(values[0])

canvas=Canvas(menu, width=600)
canvas.place(x=0, y=170)
canvas.create_line(0,5,600,5, fill="red", width=3)
canvas.create_line(300,5,300,200, fill="red", width=3)

# PHUONG THUC THANH TOAN
lbtt = Label(menu, text='Chọn phương thức thanh toán:', font=myfont)
lbtt.place(x=30, y=180)

var1=IntVar()
var1.set(0)

tienmat = Radiobutton(menu, text='Tiền mặt',font=ttfont,fg='red', value=0,variable=var1,command=color)
tienmat.place(x=20,y=210)
chuyenkhoan = Radiobutton(menu, text='Chuyển khoản (Discount 5%)',font=ttfont, value=5,variable=var1,command=color)
chuyenkhoan.place(x=20,y=250)

button = Button(menu, text='THANH TOÁN',padx=65, font = myfont, fg='red')
button.place(x=320, y = 180)
button.bind("<Button>",total)

thanhtoan=Label(menu, text='Số tiền cần thanh toán: ', font=myfont)
thanhtoan.place(x=320,y=230)

tongtien=Label(menu,text='',fg='red',font=myfont)
tongtien.place(x=320,y=260)
menu.mainloop()
```
