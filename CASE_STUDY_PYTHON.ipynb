```python
from tkinter import *
from time import sleep
from random import randint
#from playsound import playsound

img=[0,0,0,0]
y=-20
x=randint(10, 580)

game = Tk()
game.title('Catch Candy')
canvas = Canvas(game, width=600, height=450, bg='white')
canvas.pack()

img[0] = PhotoImage(file='bgcandy.png')
img[1] = PhotoImage(file='bowl128.png')
img[2] = PhotoImage(file='candy64.png')

background = canvas.create_image(0,0, anchor=NW, image=img[0])
bowl = canvas.create_image(0,290, anchor=NW, image=img[1])
candy = canvas.create_image(x,y, anchor=NW, image=img[2])
canvas.update()

score = 0
lose = 5
text_score = canvas.create_text(530,30, text='SCORE: '+ str(score), fill='red', font=('Times',20))
text_lose = canvas.create_text(530,60, text='LOSE: '+ str(lose),fill='red',font=('Helvatica',18))
game_over = canvas.create_text(330, 150, text='', fill='red', font=('Helvatica', 40))


def exit_game():
    game.destroy()
    canvas.update()

def try_again():
    canvas.itemconfig(game_over, text='')
    canvas.update()


def candyfall():
    global candy, score, lose
    canvas.move(candy,0,10)
    if canvas.coords(candy)[1]>360:
        canvas.delete(candy)
        y = -20
        x = randint(10, 580)
        candy = canvas.create_image(x, y, anchor=NW, image=img[2])
        lose -= 1
        canvas.itemconfig(text_lose, text='LOSE: ' + str(lose))
        if lose==0:
            gameOver=True
            game_over = canvas.create_text(330, 150, text='GAME OVER!!!', fill='red', font=('Helvatica', 40))
            canvas.itemconfig(game_over, text='GAME OVER!!!')
            canvas.delete(text_lose)
            canvas.delete(text_score)

            tryagain = Button(canvas, text='Try Again', font=('Helvatica',15), command=try_again)
            tryagain.place(x=350, y=200)

            exitgame = Button(canvas, text='Exit', font=('Helvatica',15),padx=20, command=exit_game)
            exitgame.place(x=200, y=200)



    if (canvas.coords(candy)[0]>=canvas.coords(bowl)[0]-30 and canvas.coords(candy)[0]+64<=canvas.coords(bowl)[0]+160) and\
            (canvas.coords(candy)[1]+64>=canvas.coords(bowl)[1] and
             canvas.coords(candy)[1]+64<=canvas.coords(bowl)[1]+64):
        #playsound('vacham.wav')
        canvas.delete(candy)
        y = -20
        x = randint(10, 580)
        candy = canvas.create_image(x, y, anchor=NW, image=img[2])

        score+=1
        canvas.itemconfig(text_score, text='SCORE: '+ str(score))
    canvas.update()

def right():
    global bowl
    if canvas.coords(bowl)[0]<530:
        canvas.move(bowl,20,0)
    canvas.update()

def left():
    global bowl
    if canvas.coords(bowl)[0]>0:
        canvas.move(bowl,-20,0)
    canvas.update()

def keyPress(event):
    if event.keysym == 'Right':
        right()
    if event.keysym == 'Left':
        left()

canvas.bind_all('<KeyPress>', keyPress)

gameOver=False
while not gameOver:
    candyfall()
    sleep(0.05)

game.mainloop()
```
